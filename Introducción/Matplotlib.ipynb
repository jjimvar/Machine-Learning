#### Nube de puntos

```python
#Se importa la función rand del módulo numpy.random. Así puedes usar directamente rand() para generar números aleatorios sin tener que escribir numpy.random.rand(). No se importa todo el módulo, solo la función rand.
from numpy.random import rand

x,y=rand(2,100)
z=x

plt.scatter(x,y)
plt.plot(x,z, "red") #Rpresentamos el modelo a utilizar (lineal)
plt.title("NUBE DE PUNTOS")
plt.show()
```

#### Histogramas

```python
data1=[1, 3 , 3, 3.5, 4.1, 5.01, 6.9, 6.99, 7.47]
data2= data1*3

plt.figure(figsize=(6,9))

plt.subplot(211)
plt.hist(data1, bins=10, rwidth=1.2) #bins (numero de barras), rwidth(anchura)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")

plt.subplot(212)
plt.hist(data2, bins=10, rwidth=1.2) #bins (numero de barras), rwidth(anchura)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

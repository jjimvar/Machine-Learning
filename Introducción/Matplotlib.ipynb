#### Nube de puntos

```python
#Se importa la funci칩n rand del m칩dulo numpy.random. As칤 puedes usar directamente rand() para generar n칰meros aleatorios sin tener que escribir numpy.random.rand(). No se importa todo el m칩dulo, solo la funci칩n rand.
from numpy.random import rand

x,y=rand(2,100)
z=x

plt.scatter(x,y)
plt.plot(x,z, "red") #Rpresentamos el modelo a utilizar (lineal)
plt.title("NUBE DE PUNTOS")
plt.show()
```

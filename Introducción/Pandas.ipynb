#### 0.1. Series

```python
#Series
altura = {"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}
a=pd.Series(altura)
b=pd.Series(altura, index=["Pedro", "Julia"])
print("Dataframe a partir de diccionario:\n", a)
print("---------------------------")
print("Dataframe filtado por el índice del diccionario:\n", b)
print("---------------------------")
```

```python
#Series con un mismo valor:
c=pd.Series([10,20,30], index=["test1", "test2", "test3"])
print("Dataframe con un mismo valor:\n", c)
print("---------------------------")    
print("Accedemos a un valor específico por índice:\n", c["test2"]) #De la forma estándar c.loc["test2"]
print("---------------------------")
print("Accedemos a través del rango que ocupa:\n", c[1]) #De la forma estándar c.iloc[1]
print("---------------------------")
print("Accedemos a los dos primero valores:\n", c.iloc[0:2]) #De la forma estándar c.iloc[1]
```

```python
import numpy as np
#Aplicamos funciones de numpy a la serie
d=pd.Series([1,2,3,4,5])
np.sum(d)
np.mean(d)
np.std(d)
np.square(d)
```

##### Evaluación expresiones de un df

```python
# Creación de un DataFrame inicializándolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas)
df
```

```python
#Dividimos la columna altura entre 2
df.eval("altura/2")
print("Dividimos la columna altura entre 2:\n", df.eval("altura/2"))
print("---------------------------")
#Lo añadimos como una nueva columna
df.eval("media_altura=altura/2", inplace=True)
print("Añadimos la media de la altura como una nueva columna:\n", df)
```

```python
# Evaluar una función utilizando una variable local
max_altura=180
df.eval("altura>@max_altura")
```

```python
# Aplicar una función externa a una columna del DataFrame
def func(x):
    return x+2

df["peso"].apply(func)
```

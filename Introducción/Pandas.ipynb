## 4. Agrupación

```python
#Nuevo df
info= {"Empresa": ["B", "A", "A", "C", "B", "C"],
       "Localización": ["Madrid", "Barcelona", "Madrid", "Valencia", "Barcelona", "Madrid"],
       "Ventas": [200, 150, 300, 400, 250, 500],
       "Gastos": [50, 80, 100, 70, 60, 120]}

data=pd.DataFrame(info)
data
```

```python
# Calculamos la media de las ventas por localización
data.groupby("Localización").agg({"Ventas": "mean"})
```

```python
# Calculamos la media de las ventas por localización y empresa
data.groupby(["Empresa", "Localización"]).agg({"Ventas": "mean"})
```

```python
# Calculamos la media de las ventas y gastos por localización y empresa
data.groupby(["Empresa", "Localización"]).agg({"Ventas": "mean", "Gastos": "mean"}) #También se puede especificar otras funciones como "sum", "max", "min", "std", "var", "count"...
```

## 8. Evaluación expresiones de un df

```python
# Creación de un DataFrame inicializándolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas)
df
```

```python
#Dividimos la columna altura entre 2
df.eval("altura/2")
print("Dividimos la columna altura entre 2:\n", df.eval("altura/2"))
print("---------------------------")
#Lo añadimos como una nueva columna
df.eval("media_altura=altura/2", inplace=True)
print("Añadimos la media de la altura como una nueva columna:\n", df)
```

```python
# Evaluar una función utilizando una variable local
max_altura=180
df.eval("altura>@max_altura")
```
